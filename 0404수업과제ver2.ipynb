```python
#P1
import pandas as pd
import numpy as np

n=100

def SettingData(n) :
    x1 = np.random.randint(3, size=n) # 인자로 들어온 구간에서 random으로 size만큼 정수 반환
    x2 = np.random.uniform(low=0, high=1, size=n)
    x3 = np.random.normal(loc=0, scale=1, size=n) # 정규분포
    x4 = (np.random.normal(loc=0, scale=1, size=n) * np.random.normal(loc=0, scale=1, size=n))
    # x4 = np.random.normal(loc=0, scale=1, size=n).tolist()
    x5 = np.random.chisquare(df=3, size=n)
    
    d = np.array([x1, x2, x3, x4, x5])
    d = np.transpose(d)
    
    d = pd.DataFrame(d, columns = ['x1', 'x2', 'x3', 'x4', 'x5'])
    # columns라고 명시해줌으로써 해당 list 순서대로 칼럼들을 불러올 것임
    # 명시된 이름과 동일해야함
    return (d)

ds = SettingData(n)

#P2
def newData(dd) :
    nd = dd.loc[:, ['x1', 'x2', 'x3', 'x5']][dd.x1 != 0]
    return (nd)  

nds = newData(ds)

#P3
def varComp (x) :
    n = len(x)
    ssum = 0

    for i in range(0, n) :
        ssum = ssum + x[i]*x[i]

    m = np.mean(x)
    var = (ssum - n*m*m)/(n-1)
    
    return (var)


var = varComp(nds.x2.tolist())

#P4
def split_Data(dd) :
    nr = len(dd)
  
    # Split data 
    dd['gp'] = np.random.uniform(0, 1, n)
    trd = dd[dd.gp <= 0.9]
    ted = dd[dd.gp > 0.8]
  
    return trd, ted

trd, ted = split_Data(ds)
# tuple unpacking

#P5
from sklearn.linear_model import LinearRegression 

def reg_Pred (trd) :
    y = trd['x2'] 
    X = trd.loc[:, ['x3', 'x5']]
    
    lr = LinearRegression() 
    lr.fit(X, y) 
    
    vars = {'x3' : [2],'x5' : [3]}
    tt = pd.DataFrame(vars)
    pred = lr.predict(tt)
    
    return pred


print(reg_Pred(trd))
```
