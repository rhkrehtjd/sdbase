```r
########################################################
# P1 : 주사위를 던지는 실험에서 1이 나올 확률 1/6임을 보여라
par(mfrow=c(1,1))
# set.seed(123457)

NofTrials = 1000

ProbConv = function(n) {

  res_trials <- runif(n, 0, 0.6)
    # n,min,max

  Tag_S_F = rep(0, n)
  NofS = 0
  p = rep(0, n)
  
  for(i in 1:n) {
    Tag_S_F[i] = ifelse(res_trials[i] < 0.1, 1, 0)
    NofS = NofS + Tag_S_F[i]
    p[i] = NofS/i
  }

  return (p)
}

p = ProbConv(NofTrials)
plot(1:NofTrials, p, type="l")
abline(h=0.2, col="red")
# 수행할 때마다 그림은 다르게 그려질 것임


# P2 : 데이터에 outlier가 있으면(예: 3sd) 제거하고 평균 계산

data = c(2, 20, 22, 18, 25, 22, 24, 26, 21, 22, 24, 22, 25)
mean(data)

MeanWithoutOutliers = function(x) {
  outlier.index = (x > mean(x) + 3*sd(x)) | (x < mean(x) - 3*sd(x))
  x.new = x[!outlier.index] # 이상치 아닌 것만 뽑는 것 !!!
  
  res = mean(x.new)
  
  return(res)
}

MeanWithoutOutliers(data)


# P3 : 확률분포 난수를 이용한 데이터 셋 생성 (범주형 변수 포함)
n = 1000

SettingData = function(n){

  d1 = rbinom(n, size=1, prob=0.7)
  d2 = runif(n, min=0, max=1)
  d3 = rnorm(n, mean=0, sd=1)
  d4 = rchisq(n, df=3)

  return (data.frame(d1, d2, d3, d4))  
}

data = SettingData(n)

```
